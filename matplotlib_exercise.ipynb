```python
# 使うライブラリ
# !pip install japanize-matplotlib

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import japanize_matplotlib
japanize_matplotlib.japanize()
```

## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df_hist_data = pd.read_csv('data/hist_data.csv')
plt.xlim(0, 54)             
plt.ylim(0, 29)                
plt.title("aデータとbデータの温度に関する度数分布", fontsize=14)  
plt.xlabel("温度(℃)", fontsize=13)     
plt.ylabel("度数", fontsize=13)     
plt.grid(False)                      
plt.tick_params(labelsize = 12)   

plt.hist(df_hist_data['aデータ'], alpha=0.4, color='#ff7f00' ,bins=30, label="aデータ")
plt.hist(df_hist_data['bデータ'], alpha=0.4, color='c',bins=30, label="bデータ")
plt.legend(loc='upper right')
plt.show()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df_scatter_data = pd.read_csv("data/scatter_data.csv")
plt.title('クラス0～2のlengthとwidthに関する散布図')
plt.grid(True)
plt.xlabel("length(cm)")
plt.ylabel("width(cm)")

class0 = df_scatter_data.loc[df_scatter_data['class'] == 0]
class1 = df_scatter_data.loc[df_scatter_data['class'] == 1]
class2 = df_scatter_data.loc[df_scatter_data['class'] == 2]

plt.scatter(class0["length (cm)"],class0["width (cm)"],label="class0",color="r")
plt.scatter(class1["length (cm)"],class1["width (cm)"],label="class1",color="b")
plt.scatter(class2["length (cm)"],class2["width (cm)"],label="class2",color="g")

plt.legend(loc="upper right")
plt.show
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df_line_data = pd.read_csv('data/line_data.csv')


plt.xlabel('期間')
plt.ylabel('件数')
plt.grid(True)
plt.title('期間ごとの件数の推移')
plt.plot(df_line_data['期間'],df_line_data['件数'],linewidth=1.5,marker="o")
plt.xticks(rotation=45,fontsize=10)
plt.yticks(np.arange(105, 145, 5))
plt.show
```

```python
plt.xlabel('期間')
plt.ylabel('費用(円)')
plt.grid(True)
plt.title('期間ごとの費用の推移')
plt.bar(df_line_data['期間'],df_line_data['費用'])
plt.xticks(rotation=45,fontsize=10)
plt.yticks(np.arange(0, 600000, 100000))
plt.show
```
